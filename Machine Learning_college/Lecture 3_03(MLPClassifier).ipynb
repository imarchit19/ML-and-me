```python
from sklearn.neural_network import MLPClassifier
import sklearn.datasets as ds
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
clf=MLPClassifier()  # creating object 
iris=ds.load_iris()  # loading dataset
X=iris.data
Y=iris.target
xtrain,xtest,ytrain,ytest=train_test_split(X,Y)
clf.fit(xtrain,ytrain) #training neural network 
```

```python
clf.score(xtest,ytest) #obtaining score 
```

```python
clf=MLPClassifier(hidden_layer_sizes=(20,),max_iter=3000)
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest) #obtaining score 
```

```python
clf.coefs_[0].shape
```

```python
clf.coefs_[1].shape
```

```python
clf=MLPClassifier(solver='lbfgs', alpha=0.0001,hidden_layer_sizes=(20, ), random_state=1)
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```

```python
import matplotlib.pyplot as plt
plt.plot(xtest,ytest)
plt.show()
```

```python
clf=MLPClassifier(alpha = 0.0002, hidden_layer_sizes=(100, ), random_state=1, )
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```

```python
clf=MLPClassifier(solver='lbfgs', alpha=10e-13,hidden_layer_sizes=(20, ), random_state=1, learning_rate = 'adaptive')
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```

```python
clf=MLPClassifier(activation = 'logistic', solver='sgd', alpha=0.0001,hidden_layer_sizes=(20, ), random_state=1, power_t = 'double')
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```

```python
clf=MLPClassifier(activation = 'tanh', solver='lbfgs', alpha=1e-9,hidden_layer_sizes=(120, ), random_state=2,beta_2 =0.888, power_t = 'double', warm_start=True)
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```

```python
clf.coefs_[0].shape
```

```python
clf.coefs_[1].shape
```

```python
import matplotlib.pyplot as plt
plt.plot(xtest,ytest)
plt.show()
```

```python
import matplotlib.pyplot as plt
pred = clf.predict(xtest)
plt.plot(pred,ytest)
plt.show()
print(pred - ytest)
```
