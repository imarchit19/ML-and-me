```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('xclara1.csv')
```

```python
dataset.head()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(dataset)
kmeans.fit_predict(dataset) 
```

```python

plt.figure(figsize=(12,6))
plt.scatter(dataset['V1'], dataset['V2'], c='b', s=30)
```

```python
plt.scatter(dataset['V1'], dataset['V2'], c=kmeans.labels_)
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(dataset['V1'], dataset['V2'], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0],kmeans.cluster_centers_[:,1], marker='*', s=300, c='r')
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
da = datasets.load_diabetes()
```

```python
x = da.data
df=pd.DataFrame(x)
```

```python
df.head()
```

```python
new = df.drop([6, 7, 8, 9], axis=1)
```

```python
new.head(5)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(new)
kmeans.fit_predict(new) 
```

```python
plt.figure(figsize=(12,6))
plt.scatter(new[0], new[1], c='b', s=30)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(new[2], new[3], c='b', s=30)
```

```python
plt.scatter(new[2], new[3], c=kmeans.labels_)
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(new[2], new[3], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 2],kmeans.cluster_centers_[:,1], marker='*', s=300, c='r')
```
