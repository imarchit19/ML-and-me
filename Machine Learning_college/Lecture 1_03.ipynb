##### 1) Armstrong Number

```python
num = int(input("Enter a number "))
count = 0
temp = num
while (temp > 0):
   letter = temp % 10
   count = count + letter ** 3
   temp = temp // 10

if (num == count):
   print("It is an Armstrong number")
else:
   print("Not an Armstrong number")
```

##### 2) Armstrong Number in Series

```python
def armstrong(n):
    count = 0
    temp = n
    while (temp > 0):
        letter = temp % 10
        count = count + letter ** 3
        temp = temp // 10
        
    if(n == count):
        return True
    else:
        return False

no = int(input("Enter a final number: "))
for i in range(0, no):
    if(armstrong(i) == True):
        print(i)
```

##### 3) Sum of Natural Numbers

```python
initial = int(input("Enter a initial number: "))
final = int(input("Enter a final number: "))

sum = 0

for i in range(initial, final + 1):
    sum += i
print(sum)
```

##### 4) Factorial of a number

```python
fact_num = int(input("Enter a number: "))
fact = 1

for i in range(1, fact_num + 1):
    fact *= i
print(fact)
```

##### 5) Multiplication Table

```python
mul_num = int(input("Enter multiplicative number: "))
fin_num = int(input("Enter final number: "))

for i in range(0, fin_num + 1):
    print(mul_num, "X", i, "=", mul_num * i)
```

##### Wave Form

```python
def WaveForm(a, n):  
    for i in range(0, n, 2): 
        if (i> 0 and a[i] < a[i-1]): 
            (a[i],a[i-1]) = (a[i-1],a[i]) 
        if (i < n-1 and a[i] < a[i+1]): 
            (a[i],a[i+1]) = (a[i+1],a[i]) 
  
a = [1, 2, 3, 4, 5, 6, 7] 
WaveForm(a, len(a)) 
print(a)
```
