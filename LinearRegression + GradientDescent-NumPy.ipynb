```python
import numpy as np
```

```python
def step_gradient(x, y, learning_rate, m, b):
    slope_m = 0
    slope_b = 0
    
    n = x.shape[0]
    
    for i in range(n):
        slope_m += (1/n) * (m*x[i]+b-y[i])*x[i]
        slope_b += (1/n) * (m*x[i]+b-y[i])
        
    new_m = m - learning_rate*slope_m
    new_b = b - learning_rate*slope_b
    
    return new_m, new_b
```

```python
def gradient_descent(x, y, learning_rate=0.001, init_m=0, init_b=0, n_epochs=10000):
    
    m = init_m
    b = init_b
    
    for i in range(n_epochs):
        m, b = step_gradient(x,y,learning_rate, m, b)
    
    return m,b
```
