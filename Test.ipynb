```python
import numpy as np
```

```python
'''
EQUATION WE ARE EXPECTING

Y = 3x + 5
'''

#CREATING X VALUES
X = np.arange(0,50)

#GENERATING Y FROM X USING EQUATION AND ADDING NOISE
Y = 3*X + 5 + 10*(2*np.random.rand(50) - 1)
```

```python
'''
PLOTTING THE GRAPH OF SYNTHETIC DATASET
'''

import matplotlib.pyplot as plt

#SETS FIGURE SIZE
plt.figure(figsize=(10,8))

#PUTS POINTS ON GRAPH USING X and Y
plt.scatter(X,Y)

#DISPLAYS GRAPH
plt.show()
```

## sk-learn (ML Library)

```python
#IMPORTING LINEAR REGRESSION CLASS

from sklearn.linear_model import LinearRegression
```

```python
#CREATING AN OBJECT OF LINEAR REGRESSION CLASS

clf = LinearRegression()
```

```python
#FITTING THE VALUES OF THE DATASET IN THE MODEL
'''
Reshape because sklearn takes only 2D data
'''
X, Y = X.reshape(-1,1), Y.reshape(-1,1)

clf.fit(X,Y)
```

```python
# FINDING SLOPE AND INTERCEPT 
# m and b

clf.coef_, clf.intercept_
```

```python
clf.predict([[130]])
```

```python
#ALL PREDICTED X VALUES
clf.predict(X)
```

```python
plt.figure(figsize=(10,8))

#PLOTTING THE X AND PREDICTED Y
#THE PREDICTED LINE

plt.plot(X, clf.predict(X), c='r')

#PLOTTING ACTUAL POINTS
plt.scatter(X,Y)

plt.show()
```
